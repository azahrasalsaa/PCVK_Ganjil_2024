**STEP 1**

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

**STEP 2**

```python
# Membuat list untuk menyimpan url dari beberapa image
urls  = ["https://i.pinimg.com/474x/3f/72/f8/3f72f87670a143bfaf354707925c0493.jpg",
          "https://i.pinimg.com/564x/35/66/fb/3566fb6a1401703c145e9662bc3a7958.jpg",
          "https://i.pinimg.com/236x/83/5d/1d/835d1df3e32c176fa6a3d2e458ab06f7.jpg"]
# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                          # read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) # resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  # convert color to RGB
  final_frame = cv.hconcat((image, image_2))      # concatenate image
  cv2_imshow(final_frame)                         # show image
  print('\n')
```

**TUGAS**

***garis vertikal***

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis horizontal ditengah image
for y in range (tinggi):
  image_3[y, int((lebar)/2)] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

***garis menyilang***

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis horizontal ditengah image
for y in range (tinggi):
  image_3[y, int((lebar/tinggi) * y)] = [255,255,255]
  image_3[y, lebar - 1 - int((lebar/tinggi) * y)] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

***garis horizontal tengah***

```python

image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Panjang garis yang diinginkan
panjang_garis = 150

# Hitung posisi awal dan akhir dari garis agar berada ditengah
garis_awal = int((lebar - panjang_garis) / 2)
garis_akhir = garis_awal + panjang_garis

# Membuat garis horizontal ditengah image
for y in range (garis_awal, garis_akhir):
  image_3[int(tinggi/2), y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

***kotak di tengah gambar***

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Menentukan ukuran kotak
ukuran_kotak = 100

# Menentukam posisi kotak
a = 80 # Jarak dari kiri
b = 60 # Jarak dari atas

# Menentukan batas akhir kotak
a_end = a + ukuran_kotak
b_end = b + ukuran_kotak

# Memastikan kotak tidak keluar dari batas gambar
a_end = min(a_end, lebar)
b_end = min(b_end, tinggi)

# Membuat kotak putih
image_3[b:b_end, a:a_end] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
